```python
df.dropna(inplace=True)
df.drop_duplicates(subset=['Question','Answer'])
```

```python
df['Q-length'] = df['Question'].apply(len)
df['A-length'] = df['Answer'].apply(len)
```

```python
max_len = 50
```

```python
df = df[(df['Q-length']<=max_len) & (df['A-length']<=max_len)]
df.shape
```

```python
df = df.sample(n=1000, random_state=42)
```

```python
df.Category.unique()
```

```python
df['Q-length'].max(), df['A-length'].max()
```

```python
dataset = Dataset.from_pandas(df)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, TrainingArguments, Trainer

model_name = "google/mt5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
def preprocess(example):
    input_text = "سؤال: " + example['Question']
    target_text = example['Answer']
    model_inputs = tokenizer(input_text, max_length=max_len, truncation=True, padding="max_length")
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(target_text, max_length=max_len, truncation=True, padding="max_length")
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

tokenized_dataset = dataset.map(preprocess, remove_columns=["Question", "Answer"])
```

```python
training_args = TrainingArguments(
    output_dir="./arabic_medical_chatbot",
    run_name='arabic_chatbot',
    per_device_train_batch_size=4,
    num_train_epochs=3,
    logging_steps=1,
    save_strategy="epoch"
    )

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset
)

trainer.train()
```

```python
def chatbot(question: str):
    input_text = f"سؤال: {question}"
    inputs = tokenizer(input_text, return_tensors="pt", truncation=True, padding=True)
    # Move inputs to the same device as the model
    for key in inputs:
        inputs[key] = inputs[key].to(model.device)
    output_ids = model.generate(**inputs, max_length=max_len, num_beams=4, early_stopping=True, max_new_tokens=20)
    response = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return str(response)
```

```python
def chatbot(question: str):
    input_text = f"سؤال: {question}"
    inputs = tokenizer(input_text, return_tensors="pt", truncation=True, padding=True)

    # Move to model device
    for key in inputs:
        inputs[key] = inputs[key].to(model.device)

    # Generate response (using max_new_tokens only)
    output_ids = model.generate(
        **inputs,
        max_new_tokens=20,  # increase if needed
        num_beams=4,
        early_stopping=True
    )

    # Decode and clean output
    response = tokenizer.decode(output_ids[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)

    import re
    response = re.sub(r'<extra_id_\d+>', '', response).strip()

    return response
```

```python
print(chatbot("ما هي أعراض مرض السكري؟"))
```
